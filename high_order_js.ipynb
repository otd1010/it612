```python
%%writefile code.js
// Part A
const data = [5, 12, 3, 20, 11, 9, 30];

let imperative = [];
for (let value of data)
  if (value>10)
    imperative.push(value * 2);
const declarative = data.filter(x => x > 10).map(x => x *2);
console.log(imperative.filter(val => declarative.includes(val)));


// Part B
const numbers = [1, 2, 3, 4, 5, 6];
console.log("\n" + numbers.filter(x => x % 2 === 0).map(x => x * 2).reduce((acc, x) => acc + x, 0))


// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

console.log("\n" + "Is any user under 18? " + users.some(user => user.age < 18));
console.log("Find the user with id === 3: ", users.find(user => user.id === 3));
users.forEach(user => console.log(user.name));


// Part D
function repeat(n, action) {[...Array(n)].forEach((_, i) => action(i));}
console.log("\n");
repeat(3, console.log);
repeat(4, i => console.log("Step " + i));
```
